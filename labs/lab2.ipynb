## Feedforward Neural Network

The feedforward neural network is one of the most basic artificial neural networks. In this ANN, the data or the input provided travels in a single direction. It enters into the ANN through the input layer and exits through the output layer while hidden layers may or may not exist. So the feedforward neural network has a front-propagated wave only and usually does not have backpropagation.

```python
import numpy as np
```

```python
def sigmoid(z):
    return 1.0/(1.0 + np.exp(-z))
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.array([1.0, 0.0, 1.0])

W_hidden = np.array([[0.2, -0.3],
                     [0.4,  0.1],
                     [-0.5, 0.2]])

b_hidden = np.array([-0.4, 0.2])

W_out = np.array([-0.3, -0.2])

b_out = 0.1

y_true = 1.0
```

```python
Z_hidden = x.dot(W_hidden) + b_hidden
H = relu(Z_hidden)
Z_out = H.dot(W_out) + b_out
y_hat = relu(Z_out)
```

```python
#Check for error
mse = 0.5 * (y_true - y_hat)**2
```

## Results

```python
print("Z_hidden =", Z_hidden)
print("H (ReLU) =", H)
print("Z_out =", Z_out)
print("y_hat (ReLU) =", y_hat)
print("MSE =", mse)
```

## Conclusion

In this activity, we performed a single forward pass on a simple feedforward neural network, where we computed the weighted sums of inputs and added biases for the hidden layer, applied the ReLU activation function to introduce non-linearity, and used the hidden layer outputs to calculate the final network output. We then compared the predicted output against the actual target using Mean Squared Error (MSE). This exercise demonstrated the step-by-step mechanics of how inputs are transformed into outputs in a neural network and reinforced the flow of data as Input → Weighted Sum + Bias → Activation → Output → Error Calculation, highlighting how each layer contributes to the network’s prediction and how performance is measured.
